```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline

cv_number = 3
poly_degrees = [1, 2, 3, 4]
max_iters = [1000, 1500, 2000]
penalties = ['l1', 'l2', 'none']
c_values = [1, 3, 6]
random_state = [12]
max_features = ['sqrt', 'log2']
n_estimators = [50, 100, 150]

methods_stats_df = pd.DataFrame(data=None, columns=['method', 'accuracy', 'scaler', 'params'])
methods_stats_df_index = 0

def append_method_stats(method, accuracy, scaler, params):
    global methods_stats_df_index
    methods_stats_df.loc[methods_stats_df_index] = [method, accuracy, scaler, params]
    methods_stats_df_index += 1
```

```python
# train_x_split, test_x_split, _, __ = train_test_split(train_x, train_y, test_size=0.2, random_state=random_state[0], stratify=train_y)
# train_indices = sorted(train_x_split.index)
# test_indices = sorted(test_x_split.index)
# 
# cv_indices = [(train_indices, test_indices)]
```

```python
standard_scaler_pipeline = Pipeline([('scaler', StandardScaler()),
                                     ('poly', PolynomialFeatures()),
                                     ('logreg', LogisticRegression())])

min_max_scaler_pipeline = Pipeline([('scaler', MinMaxScaler()),
                                    ('poly', PolynomialFeatures()),
                                    ('logreg', LogisticRegression())])

param_grid = {
    'poly__degree': poly_degrees,
    'logreg__solver': ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga'],
    'logreg__max_iter': max_iters,
    'logreg__penalty': penalties,
    'logreg__C': c_values,
    'logreg__random_state': random_state
}
```

```python
gridsearch = GridSearchCV(standard_scaler_pipeline,
                          param_grid=param_grid,
                          scoring='accuracy',
                          cv=cv_number)
gridsearch.fit(train_x, train_y)
```

```python
append_method_stats('LogisticRegression', gridsearch.best_score_, 'standard', gridsearch.best_params_)
```

```python
gridsearch = GridSearchCV(min_max_scaler_pipeline,
                          param_grid=param_grid,
                          scoring='accuracy',
                          cv=cv_number)
gridsearch.fit(train_x, train_y)
```

```python
append_method_stats('LogisticRegression', gridsearch.best_score_, 'minmax', gridsearch.best_params_)
```
